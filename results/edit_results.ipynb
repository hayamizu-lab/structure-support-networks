```python
import re

def extract_metrics(filename):
    """
    ファイルを読み込んで
      - Minimum level: の後の整数
      - Runtime (sec): の後の浮動小数点
    を取り出してタプルで返す。
    """
    level = None
    runtime = None
    pattern_level   = re.compile(r"^Minimum level:(\d+)")
    pattern_runtime = re.compile(r"^Runtime \(sec\):([\d\.]+)")

    with open(filename, 'r', encoding='utf-8') as f:
        for line in f:
            line = line.strip()
            m = pattern_level.match(line)
            if m:
                level = int(m.group(1))
                continue
            m = pattern_runtime.match(line)
            if m:
                runtime = float(m.group(1))
                continue

    return level, runtime

def main():
    if len(sys.argv) != 2:
        print(f"Usage: python {sys.argv[0]} <input_txt_file>")
        sys.exit(1)

    filename = sys.argv[1]
    level, runtime = extract_metrics(filename)

    if level is None or runtime is None:
        print("Error: ファイルから必要な情報を抽出できませんでした。")
        sys.exit(1)

    print(f"Minimum level: {level}")
    print(f"Runtime (sec): {runtime:.10f}")



########## データをまとめる
for r in range(1, 35):
    with open(f"LevelMinimization-Experiment/r={r}/exact_vs_heuristic_r={r}_summary.csv", "w") as f:
        f.write("Input file,Exact minimum level,Heuristic minimum level,Exact runtime,Heuristic runtime\n")
    for i in range(100):
        heuristic_path = f"LevelMinimization-Experiment/r={r}/exp2_r={r}_{i}/exp2_r={r}_{i}_heuristic_result.txt"
        exact_path = f"LevelMinimization-Experiment/r={r}/exp2_r={r}_{i}/exp2_r={r}_{i}_exact_result.txt"
        hlevel , hrun = extract_metrics(heuristic_path)
        elevel , erun = extract_metrics(exact_path)
        with open(f"LevelMinimization-Experiment/r={r}/exact_vs_heuristic_r={r}_summary.csv", "a") as f:
            f.write(f"n=8_r={r}_{i}.txt,{elevel},{hlevel},{erun},{hrun}\n")


for r in range(35, 44):
    with open(f"LevelMinimization-Experiment/r={r}/exact_vs_heuristic_r={r}_summary.csv", "w") as f:
        f.write("Input file,Exact minimum level,Heuristic minimum level,Exact runtime,Heuristic runtime\n")
    for i in range(100):
        heuristic_path = f"LevelMinimization-Experiment/r={r}/exp2_r={r}_{i}/exp2_r={r}_{i}_heuristic_result.txt"
        # exact_path = f"LevelMinimization-Experiment/r={r}/exp2_r={r}_{i}/exp2_r={r}_{i}_exact_result.txt"
        hlevel , hrun = extract_metrics(heuristic_path)
        # elevel , erun = extract_metrics(exact_path)
        with open(f"LevelMinimization-Experiment/r={r}/exact_vs_heuristic_r={r}_summary.csv", "a") as f:
            f.write(f"n=8_r={r}_{i}.txt,,{hlevel},,{hrun}\n")



```

```python
import argparse

def trim_trailing_zeros(value: str) -> str:
    # 小数点以下の末尾の '0' と、もしあれば末尾の '.' を削除
    return value.rstrip('0').rstrip('.')

def update_runtime_in_file(path: str) -> None:
    # ファイルを読み込んで全行を取得
    with open(path, 'r', encoding='utf-8') as f:
        lines = f.readlines()

    if len(lines) < 2:
        raise ValueError(f"{path} に 2 行目がありません。")

    # 2 行目が "Runtime (sec):..." になっていることを確認
    prefix = "Runtime (sec):"
    line = lines[1]
    if not line.startswith(prefix):
        raise ValueError(f'2 行目が "{prefix}" で始まりません: {line!r}')

    # コロン以降の値を取り出して末尾ゼロを削除
    original = line[len(prefix):].strip()
    trimmed = trim_trailing_zeros(original)

    # 新しい 2 行目を設定
    lines[1] = f"{prefix}{trimmed}\n"

    # 上書き保存
    with open(path, 'w', encoding='utf-8') as f:
        f.writelines(lines)

if __name__ == "__main__":
    parser = argparse.ArgumentParser(
        description="2 行目の Runtime (sec): 以下の数値から末尾のゼロを削除します"
    )
    parser.add_argument('file', help='対象のテキストファイル')
    args = parser.parse_args()

    try:
        update_runtime_in_file(args.file)
        print(f"✔️ {args.file} の 2 行目を更新しました。")
    except Exception as e:
        print(f"❌ エラー: {e}")
```
